# MATH AND STATISTICS

```python
# LEW JUN LONG
# TP059638
# 30 MAY 2022

# mean
import numpy
var = [33, 67, 89, 56, 32, 12, 11, 90, 67, 45, 37, 27, 98, 99, 56, 45, 40, 87]
x = numpy.mean (var)
print(x)
```

```python
# median
import numpy
var = [33, 67, 89, 56, 32, 12, 11, 90, 67, 45, 37, 27, 98, 99, 56, 45, 40, 87]
x = numpy.median (var)
print(x)
```

```python
# mode
from scipy import stats
var = [33, 67, 89, 56, 32, 12, 11, 90, 67, 45, 37, 27, 98, 99, 56, 45, 40, 87]
x = stats.mode(var)
print(x)
```

```python
# standard deviation
import numpy
var = [33, 67, 89, 56, 32, 12, 11, 90, 67, 45, 37, 27, 98, 99, 56, 45, 40, 87]
x = numpy.std(var)
print(x)
```

```python
# Practice question
import numpy
from scipy import stats
var = [[34,67,89,55,67,55,45,56,32,11,12],[23,11,34,56,67,90,56,56,56,33],[0,1,23,24,56]]
for i in var:
    a = numpy.mean(i)
    b = numpy.median(i)
    c = stats.mode(i)
    d = numpy.std(i)
    print("Mean:",a,"Median:",b,"Mode:",c,"Standard Deviation:", d)
```

# GRAPH

```python
# linear graph
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0,7])
ypoints = np.array([0,340])

# with line
plt.plot(xpoints, ypoints)
plt.show()

# without line
plt.plot(xpoints, ypoints, 'o')
plt.show()

# multiple lines
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
# with labels and title
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([80,85,90,95,100,105,110,115,120,125])
ypoints = np.array([240,250,260,270,280,290,300,310,320,330])

plt.plot(xpoints, ypoints)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
# Scatter plot
# used for data which more than 100 (data>100)
import matplotlib.pyplot as plt
import numpy as np

x = numpy.random.normal(5.0, 1.0, 1000)
y = numpy.random.normal(10.7, 2.0, 1000)
plt.scatter(x,y)
plt.show()
```

```python
# Bar graph/plot
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

# for vertical view
plt.bar(x, y)
plt.show()

# for horizontal view
plt.barh(x, y)
plt.show()
```

```python
# Histograms graph/plot
# used for frequency
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show()
```

```python
# Pie Charts graph/plot
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35,25,25,15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels=mylabels)
plt.legend()
plt.show()
```
